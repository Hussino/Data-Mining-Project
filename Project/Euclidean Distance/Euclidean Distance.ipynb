```python
import pandas as pd
from scipy.spatial.distance import euclidean
df = pd.read_csv("USA_cars_datasets.csv")


df['mileage'] = df['mileage'].replace(0.0, df.mileage.median())
df['price'] = df['price'].replace(0.0, df.price.median())


df.brand.unique()
df['brand'] = df['brand'].map({
    'mercedes-benz': 1,
    'bmw':2,
    'lexus':3,
    'infitini':4,
    'jaguar':5,
    'toyota':6,
    'nissan':7,
    'ford':8,
    'chevrolet':9,
    'lincoln':10,
    'maserati':11,
    'dodge':12,
    'jeep':13,
    'gmc':14,
    'honda':15,
    'audi':16,
    'chrysler':17,
    'mazda':18,
    'hyundai':19,
    'kia':20,
    'ram':21,
    'land':22,
    'peterbilt':23,
    'harley-division':24,
    'buick':25,
    'cadillac':26,
    'heartland':27,
    'acura':28
})
df.head()
```

```python
features = ['mileage', 'brand', 'price', 'year']
df = df[features]
df.head(15)
df.dropna(inplace=True)
# df.max()
```

```python
perfect_car = {'mileage': 5000, 'brand': 1, 'price':3000, 'year':2020}
perfect_car = pd.DataFrame([perfect_car])

def inverse_distance_similarity(carA, carB):
    euclidean_dist = euclidean(carA, carB)
    similarity = 1 / (1 + euclidean_dist)
    return similarity

distances = []
similarities = []
for i in range(len(df)):
    distance = euclidean(df[i:i+1], perfect_car)
    distances.append(distance)
    similarities.append(inverse_distance_similarity(df[i:i+1], perfect_car))


df['distance'] = distances
df['similarity'] = similarities

sorted_df = df.sort_values(by='similarity', ascending=False)

print(sorted_df)
```
